##importing nltk

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
stop_words = stopwords.words('english')
len(stop_words)
```

```python
df['stop_words']=df['reviews'].apply(lambda x: len([word for word in x.split() if word.lower() in stop_words]) )
```

```python
df.head()
```

```python
from nltk.corpus import stopwords
df['stop_words_rate'] = df['stop_words'] / df['count_word']
df
```

## Data Cleaning

```python
df.reviews
```

```python
df['lower_case']=df['reviews'].apply(lambda x: " ".join(words.lower() for words in x.split()))
```

```python
from nltk.tokenize.sonority_sequencing import punctuation
df['punctuation']=df['lower_case'].str.replace('[^\w\s]', '')
```

```python
df['stopwords']=df['punctuation'].apply(lambda x: " ".join(word for word in x.split() if word not in stop_words))
```

```python
df.head()
```

```python
len(" ".join(df['stopwords']).split())
```

```python
pd.Series(" ".join(df['stopwords']).split()).value_counts()[:30]
```

```python
other_stop_words = ["us","would","got","wasnt","everything","one","even","say","whole","going"]
```

```python
df['cleaningreviews'] = df['stopwords'].apply(lambda x: " ".join(word for word in x.split() if word not in other_stop_words))
pd.Series(" ".join(df['cleaningreviews']).split()).value_counts()[:30]
```

```python
df.head()

```

## Lemmatization

```python
from textblob import Word
nltk.download('wordnet')
```

```python
df['lemmatzed'] = df['cleaningreviews'].apply(lambda x: " ".join(Word(word).lemmatize() for word in x.split()))
df.head()
```

## Sentiment Analysis

```python
from textblob import TextBlob
```

```python
df['polarity']=df['lemmatzed'].apply(lambda x: TextBlob(x).sentiment[0])
df['subjectivity']=df['lemmatzed'].apply(lambda x: TextBlob(x).sentiment[1])
```

```python
df.drop(['lower_case','punctuation','stopwords','cleaningreviews'], axis=1, inplace=True)
```
